```python
# 2 - FUNCIONES

# IMPRIMIR INSTRUCCIONES JUEGO

def instrucciones_juego():    

    instrucciones = '''INSTRUCCIONES HUNDIR LA FLOTA:

    2 jugadores: tú y la máquina.
    El juego consiste en hundir la flota del contrincante, la máquina, en nuestro caso. 
    Para ello, tendrás que adivinar las coordenadas en las que estan posicionados sus barcos.
    Cada jugador tendrá dos tableros de 10x10:
        - Tablero Barcos: tablero con 10 barcos, colocados aleatoriamente. Los barcos se representan en el tablero con una letra "O" mayuscula.
            Cada jugador tiene 10 barcos en su Tablero Barcos:
                1 Buque: barco de eslora 4 
                2 Submarinos: barco de eslora 3
                3 Cruceros: barco de eslora 2
                4 Lanchas: barcos de eslora 1
                *Eslora = número de posiciones que ocupa el barco en el tablero  
        - Tablero Disparos: tablero donde se marcarán los disparos al 'Tablero Barcos' del contrincante. 
    Cada jugador dispone de un turno de disparo que se irá alternando. 
    En su turno de disparo, deberá decir las coordenadas en las que cree que su contrincante ha colocado un barco:
        Coordenadas: (fila, columna)
            El juego le pedirá un número del 0 al 9 para seleccionar una fila, y otro numero del 0 al 9 para seleccionar la columna. 
            Con los números de fila y columna, tenemos la coordenada de disparo.
            Si acierta, marcamos la casilla en el tablero disparos con una letra "X" que representa el barco, o la parte del barco, que estaba en esa casilla.
            Si acierta, puede volver a disparar.
            Si falla, marcamos la casilla en el tablero disparos con una letra "-", que representa el disparo en el agua.
            Si falla, es el turno de disparo de su contrincante.
    Gana el jugador que antes consiga hundir la flota del otro.
    '''
    print(instrucciones)

# CREAR TABLERO:

def crear_tablero(dimensiones, relleno): 
    '''
    Funcion para crear tableros.
    Input:
        Dimensiones del tablero: por defecto 10x10
        Relleno: "agua" - (str: " ")
    Output:
        Tablero 10x10 relleno de "agua"/espacios en blanco --> numpy array
    '''
    return np.full(dimensiones, relleno)

# IMPRIMIR TABLEROS JUGADOR

def imprimir_tableros(tablero_disparos_jugador,tablero_barcos_jugador):
    '''
    Funcion para imprimir los tableros del jugador.
    Input:
        Tablero Barcos
        Tablero Disparos
    Output:
        La impresion en pantalla de ambos tableros  
    '''
    print(f"Tablero Disparos:\n{tablero_disparos_jugador}\nTablero Barcos:\n{tablero_barcos_jugador}")

# 3 - COLOCAR BARCOS EN TABLERO:

def coordenadas_aleatorias():
    '''
    Funcion que nos da una coordenada aleatoria dentro de un tablero 10x10
    Output:
        (fila, columna) --> tuple
            fila: numero aleatorio del 0 al 9 --> int
            columna: numero aleatorio del 0 al 9 --> int
    '''
    
    fila = np.random.randint(10)
    columna = np.random.randint(10)

    return fila, columna

def direccion_aleatoria():
    '''
    Funcion que nos da una orientacion (N, S, E u O) de forma aleatoria.
    Output:
        'N' si orientacion norte (arriba)--> str
        'S' si orientacion sur (abajo)--> str
        'E' si orientacion este (derecha) --> str
        'O' si orientacion oeste (izquierda) --> str
    '''
    orientacion = random.choice(['N', 'S', 'E', 'O'])
    return orientacion

def posicion_libre(tablero_barcos, fila_barco, columna_barco, orientacion, eslora):
    '''
    Funcion que comprueba si en la posicion dada hay agua.
    Input:
        tablero_barcos: tablero de 10 dimensiones donde se quiere posicionar los barcos --> np array
        fila_barco: n de fila donde se quiere posicionar el barco (del 0 al 9) --> int
        columna_barco: n de columna donde se quiere posicionar el barco (del 0 al 9) --> int 
        orientacion: ('N', 'S', 'E', 'O') si barco mayor a 1 de eslora, hacia donde se va a posicionar el barco --> str
        eslora: n de posiciones que va a ocupar el barco en el tablero --> int
        
    Output:
        Posicion libre --> bool - True
        Posicion ocupada --> bool - False
    '''
        
    #Coordenadas barco (segun tamaño/eslora) en el tablero

    coord_N = tablero_barcos[fila_barco:fila_barco - eslora:-1, columna_barco] 
    coord_S = tablero_barcos[fila_barco: fila_barco+ eslora, columna_barco]
    coord_E = tablero_barcos[fila_barco, columna_barco: columna_barco + eslora]
    coord_O = tablero_barcos[fila_barco, columna_barco: columna_barco - eslora:-1]

    #Comprobamos coordenadas con direccion

    if orientacion == "N":

        if 0 <= fila_barco - eslora < 10 and BARCO not in coord_N:
            return True
        else:
            return False

    elif orientacion == "E":

        if 0 <= columna_barco + eslora < 10 and BARCO not in coord_E:
            return True
        else:
            return False

    elif orientacion == "S":

        if 0 <= fila_barco + eslora < 10 and BARCO not in coord_S:
            return True
        else:
            return False

    else: 
    #orientacion_barco == "O":
        if 0 <= columna_barco - eslora < 10 and BARCO not in coord_O:
            return True
        else:
            return False

def colocar_barcos(tablero_barcos, lista_barcos, BARCO):
    '''
    Funcion para colocar barcos en un tablero.
    Input:
        tablero_barcos: tablero de 10x10 donde colocar los barcos --> np array
        lista_barcos: los barcos a colocar, lista con la eslora(int) de cada barco --> list
        BARCO: representacion del barco (dibujo: O) --> str

    Output: 
        tablero_barcos con los 10 barcos colocados (20 posiciones del tablero marcadas con el str: 'O', resto agua: ' ') --> np array
    '''
    #Para cada barco de la lista:
    for eslora in lista_barcos:

        #Iniciamos vble contador -los barcos que vamos posicionando en el tablero-
        n_barcos_tablero = 0

        #Mientras no haya 10 barcos colocados en el tablero:
        while n_barcos_tablero != N_BARCOS:
            
        #Coordenadas y direccion Barco
            coordenadas_barco = coordenadas_aleatorias()    #pedimos unas coordenadas aleatorias
            fila_barco = coordenadas_barco[0]
            columna_barco = coordenadas_barco[1]
            orientacion_barco = direccion_aleatoria()       #pedimos una direccion aleatoria

            #Comprobamos que la posicion dada esté vacía y dentro del tablero
            if posicion_libre(tablero_barcos, fila_barco, columna_barco, orientacion_barco, eslora) == True:

                #añadimos un barco a la variable contador
                n_barcos_tablero += 1

                #colocamos barco en tablero_barcos
                if orientacion_barco == "N":
                    tablero_barcos[fila_barco:fila_barco-eslora:-1, columna_barco] = BARCO
                    
                elif orientacion_barco == "S":
                    tablero_barcos[fila_barco:fila_barco + eslora, columna_barco] = BARCO
                    
                elif orientacion_barco == "E":
                    tablero_barcos[fila_barco, columna_barco: columna_barco + eslora] = BARCO
                    
                else: #orientacion "O"
                    tablero_barcos[fila_barco, columna_barco: columna_barco - eslora:-1] = BARCO
                    
                break        
    #Nos devuelve el tablero_barcos con los barcos colocados            
    return tablero_barcos

def disparar():
    '''
    Funcion que pide al jugador un punto en el tablero 10x10 para disparar.
    Input:
        fila: input usuario --> int del 0 al 9
        columna: input usuario --> int del 0 al 9
    Output:
        (fila, columna)
        tupla de n enteros entre el 0 y el 9 
    '''

    #Si el usuario introduce un numero entero:
    try:

        #Pide al usuario un numero de fila del 0 al 9
        while True:
            fila_disparo = int(input("Introduzca un n de fila, del 0 al 9: "))
            #si el usuario introduce un n del 0 al 9, salimos del bucle
            if fila_disparo < 10 and fila_disparo >= 0:
                break
            #En caso de que no lo introduzca, se lo vuelve a pedir
            else:
                continue

        #lo mismo, pero para el n de la columna
        while True:
            columna_disparo = int(input("Introduzca un n de columna, del 0 al 9: "))
            
            if columna_disparo < 10 and columna_disparo >= 0:
                break
            else:
                continue

    #en caso de que el usuario no introduzca un n entero, imprimimos un mensaje de error y volvemos a llamar a la funcion
    ################ Ver esto porque nos daria un error en el juego ???¿¿
    except ValueError:
        print("No ha introducido un numero entero. Por favor, vuelva a intentarlo")
        disparar()
        
    #la funcion nos devuelve una coordenada (fila, columna) dentro del tablero 10x10 para disparar
    return fila_disparo, columna_disparo


############################################################
# A PARTIR DE AQUI NO OK:

#Que se me ocurra, falta la funcion marcar_tablero(), planteada en la celda de abajo -- pero sin tu codigo no sabía muy bien como plantearla. Podría ser algo así, pero pendiente de definir con tu codigo:

def marcar_tablero(tablero_barcos, tablero_disparos):
    '''
    Funcion para marcar los disparos en el tablero disparos y contar el numero de barcos hundidos.
    Input:
        Tablero barcos contrincante --> np array
        Tablero disparos --> np array
    '''
    #contador del n de posiciones acertadas:
    n_barcos_disparados = 0

    #tablero barcos es el de la maquina, si estamos viendo los disparos del jugador (y viceversa)
    if tablero_barcos[fila_disparo, columna_disparo] == "0":
        n_barcos_disparados += 1
        #y lo marcamos en el tablero de disparos de quien dispara
        tablero_disparos[fila_disparo, columna_disparo] = "X"
        print("¡Le has dado a un barco!")
        #vuelves a disparar

    elif tablero_barcos[fila_disparo, columna_disparo] == "X" or tablero_barcos[fila_disparo, columna_disparo] == "-":
        print("Ya habías disparado a esa coordenada")
        #turno del otro jugador


    elif tablero_barcos[fila_disparo, columna_disparo] == " ":
        tablero_disparos[fila_disparo, columna_disparo] = "-"
        print("¡Has disparado al agua!")
        #turno del otro jugador

    
```
